```python
def cleaned_startup_name(name):
  if pd.isna(name):
    return name
  name=name.strip().lower()
  if name.startswith("https://www."):
    name=name[len("https://www."):]
  elif name.startswith("https://"):
    name = name[len("https://"):]

  if name.startswith("www."):
    name=name[len("www."):]


  name = name.strip('/')

  for ext in ['.com', '.in', '.org', '.net']:
        if name.endswith(ext):
            name = name[:-len(ext)]
  name = name.split('.')[0]
  return name.capitalize()


```

```python
df2['startup_cleaned'] = df2['startup'].apply(cleaned_startup_name)
```
